```python
# Configura o pandas para exibir todas as colunas e o conteúdo completo ao inspecionar
import pandas as pd
import ast

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

```python
# Carrega o dataset bruto; "types" foi salvo como texto e volta a ser lista de strings
df = pd.read_csv("../data/pokemon_raw.csv")
df["types"] = df["types"].apply(ast.literal_eval)

print(f"Shape: {df.shape}")
print(f"Pokémons com evolution_chain None: {df['evolution_chain'].isna().sum()}")
df.head()
```

```python
# Inspeciona os valores únicos de cada atributo categórico antes de traduzir e codificar
print("=== TYPES ===")
all_types = sorted(set(t for types in df["types"] for t in types))
print(all_types)

print("\n=== COLOR ===")
print(sorted(df["color"].dropna().unique()))

print("\n=== HABITAT ===")
print(sorted(df["habitat"].dropna().unique()))

print("\n=== SHAPE ===")
print(sorted(df["shape"].dropna().unique()))

print("\n=== GENERATION ===")
print(sorted(df["generation"].dropna().unique()))
```

```python
# Lista os Pokémon agrupados por forma, para conferir se a categorização faz sentido
for shape in sorted(df["shape"].dropna().unique()):
    pokemons = df[df["shape"] == shape]["name"].tolist()
    print(f"\n=== {shape} ({len(pokemons)}) ===")
    print(", ".join(pokemons))
```

```python
# Mapas de tradução: convertem os valores da PokéAPI (inglês) para os rótulos em português.
# A geração vira número inteiro (1 a 9).
types_map = {
    "bug": "inseto", "dark": "sombrio", "dragon": "dragão",
    "electric": "elétrico", "fairy": "fada", "fighting": "lutador",
    "fire": "fogo", "flying": "voador", "ghost": "fantasma",
    "grass": "planta", "ground": "terra", "ice": "gelo",
    "normal": "normal", "poison": "veneno", "psychic": "psíquico",
    "rock": "pedra", "steel": "aço", "water": "água"
}

colors_map = {
    "black": "preto", "blue": "azul", "brown": "marrom",
    "gray": "cinza", "green": "verde", "pink": "rosa",
    "purple": "roxo", "red": "vermelho", "white": "branco",
    "yellow": "amarelo"
}

habitats_map = {
    "cave": "caverna", "forest": "floresta", "grassland": "campo",
    "mountain": "montanha", "rare": "raro", "rough-terrain": "terreno_acidentado",
    "sea": "mar", "urban": "urbano", "waters-edge": "beira_dagua"
}

shapes_map = {
    "armor": "artropode", "arms": "corpo_com_braços", "ball": "esférico",
    "blob": "amorfo", "bug-wings": "inseto_alado", "fish": "aquático",
    "heads": "múltiplas_cabeças", "humanoid": "humanoide", "legs": "corpo_com_pernas",
    "quadruped": "quadrúpede", "squiggle": "serpentiforme", "tentacles": "tentáculos",
    "upright": "bípede", "wings": "alado_não_inseto"
}

generation_map = {
    "generation-i": 1, "generation-ii": 2, "generation-iii": 3,
    "generation-iv": 4, "generation-v": 5, "generation-vi": 6,
    "generation-vii": 7, "generation-viii": 8, "generation-ix": 9
}
```

```python
# Aplica os mapas de tradução às colunas categóricas e confere os valores resultantes
df["color"] = df["color"].map(colors_map)
df["habitat"] = df["habitat"].map(habitats_map)
df["shape"] = df["shape"].map(shapes_map)
df["generation"] = df["generation"].map(generation_map)
df["types"] = df["types"].apply(lambda types: [types_map[t] for t in types])

print("Valores únicos após tradução:")
print(f"\ncolor: {sorted(df['color'].dropna().unique())}")
print(f"\nhabitat: {sorted(df['habitat'].dropna().unique())}")
print(f"\nshape: {sorted(df['shape'].dropna().unique())}")
print(f"\ngeneration: {sorted(df['generation'].dropna().unique())}")
print(f"\ntypes: {sorted(set(t for types in df['types'] for t in types))}")
```

```python
# A cadeia de evolução de poochyena/mightyena veio nula da API (falhou com HTTP 502),
# então preenchemos manualmente com a estrutura esperada.
poochyena_chain = {
    "is_baby": False,
    "species": {"name": "poochyena", "url": "https://pokeapi.co/api/v2/pokemon-species/261/"},
    "evolves_to": [{
        "is_baby": False,
        "species": {"name": "mightyena", "url": "https://pokeapi.co/api/v2/pokemon-species/262/"},
        "evolution_details": [],
        "evolves_to": []
    }],
    "evolution_details": []
}

df.loc[df["name"] == "poochyena", "evolution_chain"] = str(poochyena_chain)
df.loc[df["name"] == "mightyena", "evolution_chain"] = str(poochyena_chain)

print(f"Nulos restantes: {df['evolution_chain'].isna().sum()}")
```

```python
def get_evolution_stage(chain, pokemon_name):
    """Determina o estágio evolutivo do Pokémon procurando-o na cadeia de evolução.

    Retorna "único" (não evolui), "inicial", "médio" ou "final" conforme a posição
    na cadeia, ou None se o nome não for encontrado (ex.: formas alternativas, que
    têm sufixo no nome — tratadas na versão seguinte da função).
    """
    if isinstance(chain, str):
        chain = ast.literal_eval(chain)

    # nível 1
    if chain["species"]["name"] == pokemon_name:
        return "único" if not chain["evolves_to"] else "inicial"

    # nível 2
    for evo2 in chain["evolves_to"]:
        if evo2["species"]["name"] == pokemon_name:
            return "final" if not evo2["evolves_to"] else "médio"

        # nível 3
        for evo3 in evo2["evolves_to"]:
            if evo3["species"]["name"] == pokemon_name:
                return "final"

    return None


df["estagio_evolutivo"] = df.apply(
    lambda row: get_evolution_stage(row["evolution_chain"], row["name"]), axis=1
)

print(df["estagio_evolutivo"].value_counts())
print(f"\nNulos: {df['estagio_evolutivo'].isna().sum()}")
```

```python
# Lista os Pokémon que ficaram sem estágio (são formas alternativas, tratadas a seguir)
nulos = df[df["estagio_evolutivo"].isna()][["id", "name", "evolution_chain"]]
print(nulos[["id", "name"]].to_string())
```

```python
def get_base_name(name):
    """Remove o sufixo de forma alternativa do nome (ex.: "giratina-altered" -> "giratina").

    Os Pokémon com forma alternativa não aparecem com esse nome na cadeia de evolução,
    então reduzimos ao nome-base para conseguir localizá-los.
    """
    suffixes = [
        "-normal", "-altered", "-land", "-red-striped", "-standard",
        "-male", "-incarnate", "-ordinary", "-aria", "-shield",
        "-average", "-50", "-baile", "-midday", "-solo", "-red-meteor",
        "-disguised", "-amped", "-ice", "-full-belly", "-single-strike",
        "-basculegion", "-green-plumage", "-zero", "-curly", "-two-segment",
        "-plant", "-family-of-four"
    ]
    for suffix in suffixes:
        if name.endswith(suffix):
            return name[: -len(suffix)]
    return name


def get_evolution_stage(chain, pokemon_name):
    """Versão de get_evolution_stage que também resolve formas alternativas.

    Igual à anterior, mas compara tanto o nome quanto o nome-base (sem sufixo de
    forma), eliminando os nulos que sobravam para Pokémon com formas alternativas.
    """
    if isinstance(chain, str):
        chain = ast.literal_eval(chain)

    base_name = get_base_name(pokemon_name)

    if chain["species"]["name"] in (pokemon_name, base_name):
        return "único" if not chain["evolves_to"] else "inicial"

    for evo2 in chain["evolves_to"]:
        if evo2["species"]["name"] in (pokemon_name, base_name):
            return "final" if not evo2["evolves_to"] else "médio"

        for evo3 in evo2["evolves_to"]:
            if evo3["species"]["name"] in (pokemon_name, base_name):
                return "final"

    return None


df["estagio_evolutivo"] = df.apply(
    lambda row: get_evolution_stage(row["evolution_chain"], row["name"]), axis=1
)

print(df["estagio_evolutivo"].value_counts())
print(f"\nNulos: {df['estagio_evolutivo'].isna().sum()}")
```

```python
# Visualiza a distribuição de altura e peso para escolher os cortes de tamanho/pesado
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df["height"].plot(kind="hist", bins=30, ax=axes[0], title="Distribuição de Height")
axes[0].set_xlabel("Height (decímetros)")

df["weight"].plot(kind="hist", bins=30, ax=axes[1], title="Distribuição de Weight")
axes[1].set_xlabel("Weight (hectogramas)")

plt.tight_layout()
plt.show()

print("HEIGHT:")
print(df["height"].describe())
print("\nWEIGHT:")
print(df["weight"].describe())
```

```python
# Discretiza a altura em faixas de tamanho (pequeno/médio/grande) e cria a flag "pesado"
df["tamanho"] = pd.cut(
    df["height"],
    bins=[0, 5, 15, float("inf")],
    labels=["pequeno", "médio", "grande"]
)

df["pesado"] = df["weight"] > 100

print("Distribuição de tamanho:")
print(df["tamanho"].value_counts())
print("\nDistribuição de pesado:")
print(df["pesado"].value_counts())
```

```python
# Codifica cada atributo categórico em colunas booleanas (one-hot) e monta o df final.
# Para "types" (multivalorado), explode a lista e agrega com max para marcar todos os tipos.

# tipos
types_dummies = pd.get_dummies(
    df["types"].explode(), prefix="tipo"
).groupby(level=0).max()

# cor
color_dummies = pd.get_dummies(df["color"], prefix="cor")

# habitat
habitat_dummies = pd.get_dummies(df["habitat"], prefix="habitat")

# shape
shape_dummies = pd.get_dummies(df["shape"], prefix="forma")

# size
size_dummies = pd.get_dummies(df["tamanho"], prefix="tamanho")

# evolution_stage
stage_dummies = pd.get_dummies(df["estagio_evolutivo"], prefix="estagio")

# geração
generation_dummies = pd.get_dummies(df["generation"], prefix="geracao")

# concatena tudo
df_processed = pd.concat([
    df[["id", "name", "pesado"]],
    types_dummies,
    color_dummies,
    habitat_dummies,
    shape_dummies,
    size_dummies,
    stage_dummies,
    generation_dummies,
], axis=1)

print(f"Shape final: {df_processed.shape}")
print(f"\nColunas: {list(df_processed.columns)}")
df_processed.head()
```

```python
# Verifica Pokémon com features idênticas (os modelos não conseguem distingui-los).
# duplicatas sem considerar id e nome
cols_sem_id_nome = [c for c in df_processed.columns if c not in ["id", "name"]]
dup1 = df_processed[df_processed.duplicated(subset=cols_sem_id_nome, keep=False)]
print(f"Duplicatas sem considerar id e nome: {len(dup1)}")
if len(dup1) > 0:
    print(dup1[["id", "name"]].to_string())
```

```python
# Mesma checagem ignorando a geração: mostra quantos colidem só pelos atributos físicos
# duplicatas sem considerar id, nome e geração
cols_sem_id_nome_geracao = [c for c in df_processed.columns if c not in ["id", "name"] and not c.startswith("geracao_")]
dup2 = df_processed[df_processed.duplicated(subset=cols_sem_id_nome_geracao, keep=False)]
print(f"\nDuplicatas sem considerar id, nome e geração: {len(dup2)}")
if len(dup2) > 0:
    print(dup2[["id", "name"]].to_string())
```

```python
# Salva o dataset processado, que será a entrada do treino dos modelos (notebook 03)
df_processed.to_csv("../data/pokemon_processed.csv", index=False)
print(f"Dataset salvo! Shape: {df_processed.shape}")
```

```python
# Validação: amostra 2 Pokémon de cada geração e imprime as features ativas de cada um
feature_cols = [c for c in df_processed.columns if c not in ["id", "name"]]

geracao_cols = [c for c in df_processed.columns if c.startswith("geracao_")]
geracao_label = df_processed[geracao_cols].idxmax(axis=1)

samples = (
    df_processed.groupby(geracao_label)
    .apply(lambda g: g.sample(min(2, len(g)), random_state=42))
    .reset_index(drop=True)
)

print(f"Total de amostras: {len(samples)}\n")

for _, row in samples.iterrows():
    true_cols = [col for col in feature_cols if row[col] == True]
    print(f"#{int(row['id'])} {row['name']}")
    print(f"  {', '.join(true_cols)}\n")
```
