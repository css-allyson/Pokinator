```python
# Bibliotecas para treinar os modelos (sklearn) e salvá-los em disco (joblib)
import pandas as pd
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
pd.set_option("display.max_columns", None)
import warnings
warnings.filterwarnings("ignore", message="The number of unique classes is greater than 50%")
```

```python
# Carrega os dados processados e separa features (X) do alvo (y = nome do Pokémon)
df = pd.read_csv("../data/pokemon_processed.csv")

X = df.drop(columns=["id", "name"])
y = df["name"]

print(f"Features: {X.shape[1]}")
print(f"Pokémons: {y.nunique()}")
```

```python
# Treina a Árvore de Decisão com todos os Pokémon e salva o modelo.
# (o aviso do sklearn é esperado: há uma classe por Pokémon, não é regressão)
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X, y)

joblib.dump(dt, "../models/decision_tree.pkl")
print("Árvore de Decisão treinada e salva!")
print(f"Profundidade da árvore: {dt.get_depth()}")
print(f"Número de folhas: {dt.get_n_leaves()}")
```

```python
# Treina o Naive Bayes (gaussiano) com os mesmos dados e salva o modelo
nb = GaussianNB()
nb.fit(X, y)

joblib.dump(nb, "../models/naive_bayes.pkl")
print("Naive Bayes treinado e salvo!")
```

```python
# salva também as colunas de features para usar no backend
joblib.dump(list(X.columns), "../models/feature_columns.pkl")
print(f"Colunas salvas: {list(X.columns)}")
```
